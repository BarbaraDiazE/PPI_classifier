Filter metrics to indentify best models for each algorithm

```python
import pandas as pd
```

```python
#SVM
```

```python
SVM = pd.read_csv("SVM_summary_metrics.csv", index_col = "Unnamed: 0")
```

```python
#data.head()
```

```python
SVM[ (SVM["Precision"]>0.90) & (SVM["F1"]>0.9) & (SVM["Recall"]>0.94)]#.sort_values("id_model")
```

```python
SVM[SVM["id_model" ]== "SVM25"]
```

RF

```python
RF = pd.read_csv("RF_summary_metrics.csv", index_col = "Unnamed: 0")
```

```python
RF[ (RF["Precision"]>0.92) & 
     (RF["F1"]>0.9) & 
     (RF["Recall"]>0.93)].sort_values("id_model")
```

```python
RF[RF["id_model" ]== "RF13"]
```

LRG

```python
LRG = pd.read_csv("LRG_summary_metrics.csv", index_col = "Unnamed: 0")
```

```python
LRG[ (LRG["Precision"]>0.90) & 
     (LRG["F1"]>0.9) & 
     (LRG["Recall"]>0.92)].sort_values("id_model")
```

```python
LRG[LRG["id_model" ]== "LRG32"]
```

```python
#extract info
```

```python
df = pd.DataFrame() 
#df2 = pd.DataFrame() 
#df3 = pd.DataFrame() 
```

```python
best_rf = ["RF1  ", "RF5  ", "RF7  ", "RF12", "RF13"]
for i in best_rf:
    _ = RF[RF['id_model'] == i]
    df = df.append(_)   
df
```

```python
best_lrg = ["LRG31", "LRG32", "LRG33"]
for i in best_lrg:
    _ = LRG[LRG['id_model'] == i]
    df = df.append(_)   
df
```

```python
best_svm = ["SVM9  ", "SVM10", "SVM24", "SVM25"]
for i in best_svm:
    _ = SVM[SVM['id_model'] == i]
    df = df.append(_)   
df
```

```python
#reset index
df = df.reset_index(drop=True)
```
