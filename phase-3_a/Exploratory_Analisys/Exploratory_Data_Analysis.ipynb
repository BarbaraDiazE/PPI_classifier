```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Exploratory_Analisys:
    
    def __init__(self, input_file, Library, features):
        self.Data  = pd.read_csv(f'/home/barbara/Documents/DIFACQUIM/PPI_classifier/phase-1/Databases/{input_file}')
        self.Data = self.Data.drop("Unnamed: 0", axis = 1)
        self.Library= Library
        self.features = features
        print(str(Library), "Library contains : ", str(self.Data.shape[0]), "elements")
        
    def stats(self):
        Data = self.Data[self.features]
        stats = self.Data.describe()
        stats.to_csv((str(self.Library)+'_statistics.csv'), sep = ",")
        print(stats)
        
    def correlation(self):
        corr = self.Data.corr()
        f, ax = plt.subplots(figsize = (10,8))
        corr_matrix = sns.heatmap(corr,  annot=True)
        figure = corr_matrix.get_figure()    
        #figure.savefig((str(self.Library)+'_correlation.png'), dpi=800)
        figure.savefig('corr_tutoral.png')
        h = self.Data.hist(figsize=(10,12))
    
    def boxplot(self, feature):
        Data = self.Data
        sns.boxplot(x = "Library", y = Data[feature], data = Data, )
        plt.savefig("boxplot_feature.png")
```

```python
features = ['HBA', 'HBD', 'RB', 'LogP', 'TPSA', 'MW', 'Heavy Atom', 'Ring Count', 'Fraction CSP3']
```

```python
a = Exploratory_Analisys("Dataset.csv", "Epidatabase", features)
```
